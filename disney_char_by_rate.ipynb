```python

import operator
import math
import pandas as pd
import csv
%matplotlib inline

def popularity(selectname):
   
    sumname={}
    f = open("output/StateNames.csv")
    allStates = csv.reader(f)
    for line in allStates:
        state=line[4]
        gender=line[3]
        year=(line[2])
        name=line[1]
        count=(line[5])
        if name==selectname:
            sumname[year]=sumname.get(year,0)+int(count)
    for key,values in sorted(sumname.items(), key=operator.itemgetter(1),reverse=True): #sort by value:
        return key
    f.close()

        
Names = list(df.Name)
print(Names)
startYear=list(df.StartYear)
endYear=list(df.EndYear)

namesdict ={}
specificName={}
selectedNames =[]

for i in range(0,len(Names)):
    popyear = ''
    popyear = popularity(Names[i])
    if popyear is None:
        continue
    
    influenceStart = int(startYear[i])-1
    influenceEnd = int(endYear[i])
                           
    
    selectedNames.append(Names[i])


print(selectedNames)

f = open("output/StateNames.csv")
allStates = csv.reader(f)

for line in allStates:
        state=line[4]
        gender=line[3]
        year=(line[2])
        name=line[1]
        count=(line[5])
        if name in selectedNames:
            specificName = namesdict.get(name,{})
            specificName[year] = specificName.get(year,0)+int(count)
            namesdict[name] = specificName
            
f.close()

print(namesdict)

print("Done")
```

```python
from statistics import mean
from collections import defaultdict
avg = defaultdict(list)
avg2 = {}
avg3 ={}
for k, v in namesdict.items():
    for v1,v2 in v.items():
        avg[v1] +=[v2] 
        
    

for year,count in avg.items():
 
    avg2[year]=sum(count)

rate=defaultdict(list)
ratedic={}

for k, v in namesdict.items():
    for v1,v2 in v.items():
        rate[k] +=[(v1,v2/avg2[v1])]
for k in rate.keys():
    ratedic[k] = dict(rate[k])      
print(ratedic)
            
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
import operator
df_losing = df
#Charts    
rcParams['figure.figsize'] = 17,24
#colors = list("rgbcmykr")
f, axarr = plt.subplots(len(namesdict.items()), sharex=True)
i=0

for name, year_dict in ratedic.items():
    x=[]
    y=[]
    for key,values in sorted(year_dict.items(), key=operator.itemgetter(0)): #sort by year:
        x.append(key) #year
        y.append(values) #counts
    axarr[i].plot(x, y,label=name,linewidth=2,color='r')
    axarr[i].axvspan(int(df_losing[df_losing.Name == name].StartYear), int(df_losing[df_losing.Name == name].EndYear), alpha=0.5, color='black')
    axarr[i].set_title(name)
    i=i+1
print("done")
```
